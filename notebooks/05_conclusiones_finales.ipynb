## Proyecto ambiental – Métodos Numéricos

Este notebook presenta las conclusiones generales del análisis ambiental realizado sobre el material particulado PM2.5 en municipios de la Sabana Occidente de Cundinamarca.

Asimismo, se consolidan los resultados obtenidos mediante herramientas de análisis de datos y métodos numéricos aplicados a la ingeniería ambiental.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/datos_pm25_sabana.csv")

df.head()
```

## Tabla resumen final del proyecto

```python
tabla_final = df.groupby("municipio")[["pm25","temperatura","humedad","viento"]].mean()

tabla_final
```

## Exportación de resultados

Se exporta la tabla resumen final en formato CSV para facilitar su almacenamiento y análisis posterior.

```python
tabla_final.to_csv("../docs/resultados_finales_pm25.csv")

print("Archivo exportado correctamente")
```

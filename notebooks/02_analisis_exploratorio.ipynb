# Notebook 02: Análisis exploratorio de datos PM2.5

## Proyecto ambiental – Métodos Numéricos

En este notebook se realiza el análisis exploratorio de la base de datos ambiental correspondiente al material particulado PM2.5 en municipios de la Sabana Occidente de Cundinamarca.

El objetivo es aplicar herramientas de análisis de datos mediante pandas para identificar patrones, diferencias entre municipios, comportamiento de variables meteorológicas y posibles relaciones con la concentración de PM2.5.


```python
import pandas as pd
```

```python
df = pd.read_csv("../data/datos_pm25_sabana.csv")

df
```

## Promedio de PM2.5 por municipio

Se calcula el promedio de concentración de PM2.5 para cada municipio analizado.

```python
df.groupby("municipio")["pm25"].mean()
```

## Interpretación ambiental

Los municipios con mayores concentraciones promedio de PM2.5 presentan una mayor influencia de actividades industriales y tráfico vehicular.

Mosquera y Funza registran los valores más altos debido a la presencia de corredores industriales y logísticos. Por otro lado, Facatativá presenta menores concentraciones gracias a mejores condiciones de ventilación atmosférica y menor presión industrial.

## Temperatura promedio por municipio

```python
df.groupby("municipio")["temperatura"].mean()
```

## Humedad relativa promedio por municipio

```python
df.groupby("municipio")["humedad"].mean()
```

## Velocidad promedio del viento

```python
df.groupby("municipio")["viento"].mean()
```

## Filtrar municipios con PM2.5 alto

Se identifican registros con concentraciones superiores a 32 µg/m³.

```python
df[df["pm25"] > 32]
```

## Interpretación del filtro

Los registros con mayores concentraciones de PM2.5 corresponden principalmente a Mosquera y Funza, municipios caracterizados por una fuerte actividad industrial y tránsito de vehículos de carga.

## Aplicación de condicionales

Se crea una nueva columna para clasificar el nivel de contaminación atmosférica según la concentración de PM2.5.

```python
def clasificar_pm25(valor):
    
    if valor <= 12:
        return "Buena"
    
    elif valor <= 35:
        return "Moderada"
    
    else:
        return "Dañina"


df["clasificacion_pm25"] = df["pm25"].apply(clasificar_pm25)

df
```

## Conteo de clasificación de calidad del aire

```python
df["clasificacion_pm25"].value_counts()
```

## Tabla resumen del proyecto

Se genera una tabla resumen con los promedios ambientales por municipio.

```python
tabla_resumen = df.groupby("municipio")[["pm25","temperatura","humedad","viento"]].mean()

tabla_resumen
```

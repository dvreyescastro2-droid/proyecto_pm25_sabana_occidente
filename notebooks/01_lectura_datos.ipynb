## Método numérico utilizado

El presente informe aplica el método de **regresión por mínimos cuadrados**, propuesto por Steven C. Chapra y Raymond P. Canale en el libro *Métodos Numéricos para Ingenieros*.

Este método permite establecer relaciones matemáticas entre una variable dependiente y varias variables independientes. En este caso, se busca relacionar la concentración de PM2.5 con variables meteorológicas como temperatura, humedad relativa y velocidad del viento.

El modelo matemático aplicado corresponde a:

$$
PM_{2.5} = a_0 + a_1(T) + a_2(H) + a_3(V)
$$

Donde:

- $PM_{2.5}$ = concentración del material particulado fino.
- $T$ = temperatura ambiental.
- $H$ = humedad relativa.
- $V$ = velocidad del viento.
- $a_0, a_1, a_2, a_3$ = coeficientes del modelo.

Este modelo permite estimar el comportamiento del contaminante a partir de las condiciones ambientales observadas en los municipios analizados.

# Notebook 03: Gráficas y análisis visual del PM2.5

## Proyecto ambiental – Métodos Numéricos

En este notebook se desarrollan representaciones gráficas de las variables ambientales relacionadas con el material particulado PM2.5 en municipios de la Sabana Occidente de Cundinamarca.

Las gráficas permiten identificar patrones visuales, diferencias entre municipios y posibles relaciones entre las variables meteorológicas y la concentración del contaminante atmosférico.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/datos_pm25_sabana.csv")

df.head()
```

## Promedio de PM2.5 por municipio

```python
promedio_pm25 = df.groupby("municipio")["pm25"].mean()

promedio_pm25
```

## Gráfica de barras del PM2.5 promedio

```python
promedio_pm25.plot(kind="bar")

plt.title("Promedio de PM2.5 por municipio")
plt.xlabel("Municipio")
plt.ylabel("PM2.5 (µg/m³)")

plt.show()
```

## Interpretación ambiental

La gráfica evidencia que Mosquera y Funza presentan las mayores concentraciones promedio de PM2.5, asociadas principalmente a actividad industrial y transporte de carga.

Facatativá presenta menores niveles del contaminante debido a menores presiones industriales y mejores condiciones de ventilación atmosférica.

## Histograma de PM2.5

```python
plt.hist(df["pm25"])

plt.title("Distribución de PM2.5")
plt.xlabel("PM2.5")
plt.ylabel("Frecuencia")

plt.show()
```

## Relación entre temperatura y PM2.5

```python
plt.scatter(df["temperatura"], df["pm25"])

plt.title("Temperatura vs PM2.5")
plt.xlabel("Temperatura (°C)")
plt.ylabel("PM2.5")

plt.show()
```

## Interpretación de dispersión

La gráfica de dispersión permite visualizar la relación entre temperatura y concentración de PM2.5. Se observa que los municipios con mayores temperaturas presentan también mayores concentraciones del contaminante, especialmente en zonas industriales.

## Relación entre viento y PM2.5

```python
plt.scatter(df["viento"], df["pm25"])

plt.title("Velocidad del viento vs PM2.5")
plt.xlabel("Velocidad del viento (m/s)")
plt.ylabel("PM2.5")

plt.show()
```

## Interpretación ambiental del viento

La gráfica muestra que menores velocidades del viento favorecen la acumulación de PM2.5 en el ambiente, mientras que mayores velocidades contribuyen a dispersar las partículas contaminantes.

## Comparación de humedad relativa por municipio


```python
promedio_humedad = df.groupby("municipio")["humedad"].mean()

promedio_humedad.plot(kind="bar")

plt.title("Humedad relativa promedio")
plt.xlabel("Municipio")
plt.ylabel("Humedad (%)")

plt.show()
```

## Conclusión del análisis gráfico

Las representaciones gráficas permitieron identificar diferencias importantes en el comportamiento del PM2.5 entre los municipios evaluados.

Mosquera y Funza presentaron mayores concentraciones del contaminante debido a actividades industriales y tránsito vehicular, mientras que Facatativá mostró menores valores asociados a mejores condiciones de ventilación.

Además, las variables meteorológicas como velocidad del viento y humedad relativa demostraron influencia sobre el comportamiento atmosférico del material particulado.
